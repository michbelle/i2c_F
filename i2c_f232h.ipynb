```python
device_F232h="ftdi://ftdi:232h:1:4/1"
```

```python
slave_addr=0x69
slave_addr=0x36
slave_addr=0x68
```

```python
import time
from pyftdi.i2c import I2cController
i2c = I2cController()
i2c.configure(device_F232h)
slave = i2c.get_port(slave_addr)
while True:
    data=slave.read_from(0x47,2)
    print(data)
    gyro=int.from_bytes(data, byteorder='big', signed=False)
    print(gyro)
    time.sleep(0.001)
```
